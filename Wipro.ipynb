```python
# Write a program to check if a given number is prime or not.
num = int(input("Enter the number: "))
count = 0

for i in range(2,num+1):
    if num%i == 0:
        count+=1

if count==1:
    print("Prime no")
else:
    print("Not Prime")
```

```python
# Write a program to print prime numbers between 10 and 99.
s=int(input("enter the starting no: "))
e=int(input("enter the ending no: "))

for nums in range(s,e+1):
    count = 0

    for i in range(2,nums+1):
        if nums%i == 0:
            count+=1
    if count==1:
        print(nums)
```

```python
# Write a program to print the sum of all the digits of a given number.
# Example:
# I/P:1234
# O/P:10
num = int(input("Enter a 4-digit number: "))

sum = 0

while num > 0:
    sum += num % 10
    num = num // 10

print("Sum =", sum)
```

```python
# Write a program to reverse a given number and print.Example:1
# I/P: 1234
# O/P:4321
num = int(input("Enter a 4-digit number: "))

reverse = 0

while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num = num // 10

print("Reverse =", reverse)
```

```python
# Write a program to find if the given number is palindrome or not
# Example:1
# I/P:110011
# O/P: 110011 is a palindrome.
num = int(input("Enter a number: "))

original = num
reverse = 0

while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num = num // 10

if original == reverse:
    print(original, "is a palindrome")
else:
    print(original, "is not a palindrome")
```
